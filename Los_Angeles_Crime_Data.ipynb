```python
# Core data handling
import os
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings(
    "ignore",
    message=".*'M' is deprecated.*",
    category=FutureWarning
)
# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Make plots render inline in Jupyter
%matplotlib inline
sns.set_theme()

# Path to your data (notice the raw‑string to handle backslashes)
file_path = r"C:\Users\kyrda\OneDrive\Desktop\GITHUB Projects\Los Angeles Crime Data from 2020-2025\Crime_Data_from_2020_to_Present.csv"
print("File exists:", os.path.exists(file_path))

# Load it into a DataFrame
df = pd.read_csv(file_path)

# Peek at the first few rows
df.head()
```

```python
df.describe(include="all")
```

```python
# 1. Basic info: types and non-null counts
df.info()

# 2. Descriptive stats for all columns (numeric + categorical)
df.describe(include="all").T

# 3. Count of missing values per column
missing = df.isnull().sum()
print(missing[missing > 0])

# 4. Number of unique values per column
uniques = df.nunique()
print(uniques.sort_values(ascending=False))

# 5. (Optional) A quick per‑column summary loop
for col in df.columns:
    print(f"Column: {col}")
    print(f"  • Type:         {df[col].dtype}")
    print(f"  • Non‑null:     {df[col].count()} / {len(df)}")
    print(f"  • Unique vals:  {df[col].nunique()}")
    if df[col].dtype == "object" or df[col].nunique() < 10:
        print(f"  • Top values:\n{df[col].value_counts().head(5)}")
    print("-" * 40)
```

```python
# Build the core summary
summary = pd.DataFrame({
    'dtype':       df.dtypes,
    'non_null':    df.count(),
    'null_count':  df.isnull().sum(),
    'unique_vals': df.nunique()
})

# Functions to get “top” value and its count for object or low‑cardinality cols
def get_top(val_series, thresh=20):
    if val_series.dtype == 'object' or val_series.nunique() <= thresh:
        vc = val_series.value_counts(dropna=False)
        top_val = vc.index[0]
        top_cnt = vc.iloc[0]
        return pd.Series([top_val, top_cnt])
    else:
        return pd.Series([pd.NA, pd.NA])

summary[['top_value','top_freq']] = df.apply(get_top).T

# (Optional) reorder columns
summary = summary[['dtype','non_null','null_count','unique_vals','top_value','top_freq']]

# Display nicely
from IPython.display import display
display(summary)
```

```python
# Compute the top 5 most frequent crimes
top5 = df['Crm Cd Desc'].value_counts().nlargest(5)

# Plot
plt.figure(figsize=(6,4))
top5.plot(kind='bar')
plt.title('Top 5 Crime Categories (2020–Present)')
plt.xlabel('Crime Type')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 1. (Re)parse the date column so df exists in this cell
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')

# 3. Plot victim age distribution
plt.figure(figsize=(8, 5))
plt.hist(df['Vict Age'].dropna(), bins=30)
plt.title('Victim Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# make sure DATE OCC is datetime and indexed
daily = df.set_index('DATE OCC').resample('D').size().rename('count')

plt.figure(figsize=(10,4))
daily.plot()
plt.title('Daily Crime Reports')
plt.xlabel('Date')
plt.ylabel('Reports per Day')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.scatter(
    df['TIME OCC']//100,        # convert HHMM to hour
    df['Vict Age'],
    s=2,
    alpha=0.3
)
plt.title('Victim Age by Hour of Day')
plt.xlabel('Hour (24‑hr)')
plt.ylabel('Victim Age')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.kdeplot(df['Vict Age'].dropna(), fill=True)
plt.title('Kernel Density Estimate of Victim Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
# 2. Parse the date column
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')

# --- Aggregate 1: Monthly crime counts by crime type ---
#   (Power BI can then pivot Month × CrimeType)
monthly = (
    df
    .assign(Month=df['DATE OCC'].dt.to_period('M').dt.to_timestamp())
    .groupby(['Month', 'Crm Cd Desc'])
    .size()
    .reset_index(name='IncidentCount')
)
monthly.to_csv('monthly_crime_by_type_powerbi.csv', index=False)

# --- Aggregate 2: Total incidents by LAPD area ---
area_summary = (
    df
    .groupby('AREA NAME')
    .size()
    .reset_index(name='TotalIncidents')
)
area_summary.to_csv('crime_by_area_powerbi.csv', index=False)

# --- Aggregate 3: Average victim age by area and month ---
age_summary = (
    df.dropna(subset=['Vict Age'])
      .assign(Month=df['DATE OCC'].dt.to_period('M').dt.to_timestamp())
      .groupby(['Month', 'AREA NAME'])['Vict Age']
      .mean()
      .reset_index(name='AvgVictimAge')
)
age_summary.to_csv('avg_victim_age_by_area_month_powerbi.csv', index=False)

print("3 CSV files created:\n"
      " - monthly_crime_by_type_powerbi.csv\n"
      " - crime_by_area_powerbi.csv\n"
      " - avg_victim_age_by_area_month_powerbi.csv")
```
